```python
class BIDSmap(object):
    '''working doc-string'''
    
    def __init__(self,
                bids_search,
                bids_map):
        '''working doc-string'''
        self.bids_search = bids_search
        self.bids_map = bids_map
    
    def key_nav(self,
                key):
        '''working doc-string'''
        val_1 = self.bids_search[key]
        val_2 = self.bids_map[key]
        return {val_1:val_2}
```

```python
def get_nest_dict(dictionary,idx=0):
    '''working doc-string'''
    try:
        key_list = list(dictionary.keys())
    except AttributeError:
        return None,None
    
    if idx > (len(key_list) - 1):
        raise IndexError("Specified index is out dictionary bounds.")
        
    if isinstance(dictionary[key_list[idx]],list):
         get_nest_dict(dictionary=dictionary[key_list[idx]],idx=idx)
    else:
        return key_list[idx],dictionary[key_list[idx]]
```

```python
def check_dict(d1, d2, path="", verbose=False):
    '''
    * Use this to compare dictionaries that are supposed to be keymatched *
    
    StackOverflow answer: 
    https://stackoverflow.com/questions/27265939/comparing-python-dictionaries-and-nested-dictionaries
    '''
    for k in d1:
        if (k not in d2):
            print (path, ":")
            print (k + " as key not in d2", "\n") # raise error here
            # raise KeyError("Input dictionaries do not have matching keys")
        else:
            if type(d1[k]) is dict:
                if path == "":
                    path = k
                else:
                    path = path + "->" + k
                check_dict(d1[k],d2[k], path, verbose)
            else:
                if d1[k] != d2[k] and (d1[k] is None or d2[k] is None):
                    if verbose:
                        print (path, ":")
                        print (" - ", k," : ", d1[k])
                        print (" + ", k," : ", d2[k])
                    # raise error here
                    # raise ValueError("Input dictionaries map to NoneType values.")
```
